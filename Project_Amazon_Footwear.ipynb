```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
footwear = pd.read_csv("/content/footwear_project_data.csv")
footwear.head()
```

```python
footwear.info()
```

```python
footwear['Gender_category'] = footwear['Gender_category'].astype('category')
footwear['Category'] = footwear['Category'].astype('category')
footwear['Price (₹)'] = footwear['Price (₹)'].astype('float32')
footwear['Discount (%)'] = footwear['Discount (%)'].astype('float32')
footwear['Rating'] = footwear['Rating'].astype('float32')
footwear['Sales Count'] = footwear['Sales Count'].astype('int32')
footwear['Country'] = footwear['Country'].astype('category')
footwear['Water Resistant'] = footwear['Water Resistant'].astype('category')
footwear['Gender'] = footwear['Gender'].astype('category')
```

```python
footwear.info()
```

### Relationship between Average Discount Vs Average Sales For Top 10 Brands

```python
avg_discount = top10Brands_data.groupby('Brand')['Discount (%)'].agg('mean')
avg_discount
```

```python
avg_salescount = top10Brands_data.groupby('Brand')['Sales Count'].agg('mean')
avg_salescount
```

```python
avg_dis_salescount = pd.merge(avg_discount,avg_salescount, left_on = 'Brand', right_on = 'Brand')
avg_dis_salescount
```

```python
plt.figure(figsize=(12, 5))

# Scatter plot with distinct colors for each brand
sns.scatterplot(data=avg_dis_salescount,
                x='Sales Count',
                y='Discount (%)',
                hue='Brand',
                palette='tab10',  # 'tab10' is great for distinct colors
                s=100)            # Increases point size for clarity

# Regression line (without hue, as it shows the overall trend)
sns.regplot(data=avg_dis_salescount,
            x='Sales Count',
            y='Discount (%)',
            scatter=False,       # Avoid plotting points again
            color='black',       # Regression line in black for contrast
            line_kws={'linewidth': 2})

plt.title('Avg Sales Count vs. Avg Discount (%) by Brand')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend for better visibility
plt.savefig('Average Discount Vs Average Sales For Top 10 Brands.png')
```

### Relationship between Average Price Vs Average Sales For Top 10 Brands

```python
avg_Price = top10Brands_data.groupby('Brand')['Price (₹)'].agg('mean')
avg_Price
```

```python
avg_Price_Salescount = pd.merge(avg_Price, avg_salescount, left_on = 'Brand', right_on = 'Brand')
avg_Price_Salescount
```

```python
plt.figure(figsize=(12, 5))

# Scatter plot with distinct colors for each brand
sns.scatterplot(data=avg_Price_Salescount,
                x='Sales Count',
                y='Price (₹)',
                hue='Brand',
                palette='hsv',  # 'tab10' is great for distinct colors
                s=100)            # Increases point size for clarity

# Regression line (without hue, as it shows the overall trend)
sns.regplot(data=avg_Price_Salescount,
            x='Sales Count',
            y='Price (₹)',
            scatter=False,       # Avoid plotting points again
            color='black',       # Regression line in black for contrast
            line_kws={'linewidth': 2})

plt.title('Avg Sales Count vs. Avg Price (₹) by Brand')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend for better visibility
plt.show()
plt.savefig('Average Price Vs Average Sales For Top 10 Brands.png')
```
